```python
import numpy
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D

#
p0 = [0,6,2,1]
p1 = [-2,1,-1,1]

theta = numpy.linspace(0, numpy.pi, 100)
a0 = numpy.cos(theta)
a1 = numpy.sin(theta)

w = a0 + a1
x = a0 * p0[0] + a1 * p1[0]
y = a0 * p0[1] + a1 * p1[1]
z = a0 * p0[2] + a1 * p1[2]

fig = pyplot.figure()
ax = fig.add_axes(Axes3D(fig))

ax.plot(x/w, y/w, z/w, label = 'line generated by homogenous coordinates')

ax.legend()

pyplot.show()
```
